```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4])
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro') # third argument red circle
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0,6,0,20]) # x-min x-max y-min y-max
plt.show()
```

```python
import numpy as np
t = np.arange(0.,7.,0.5)
plt.plot(t,t,'gs',t,t**2,'r^',t,t**3,'k:')
plt.axis([0,10,0,400])
plt.show()
```

```python
data = {
    'a': [1,2,3,4],
    'b': [10, 22, 15, 33]
}
plt.scatter('a', 'b', c='c', data=data)
plt.xlabel('a-side on x-axis')
plt.ylabel('b-side on y-axis')
plt.show()
```
